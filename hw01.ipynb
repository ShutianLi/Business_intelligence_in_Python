```python
import pandas as pd 
data_link = '/Users/user/Desktop/Business_intelligence_in_Python/ToyotaCorolla.csv'
toyota_df = pd.read_csv(data_link,encoding = 'unicode_escape')
```

```python
print(toyota_df.index)
toyota_df.head()
```

```python
toyota2_df = toyota_df.drop(toyota_df.index[2:199],0,inplace = False)
toyota2_df.head()
```

```python
toyota2_df.drop(toyota2_df.index[499:501],0,inplace=True)
toyota2_df.describe()
```

```python
toyota2_df.drop(toyota2_df.index[999:1204],0,inplace=True)
toyota2_df.describe()
```

```python
toyota2_df.drop('Met_Color',axis=1,inplace = True)
toyota2_df.describe()
```

```python
toyota2_df.loc[toyota_df.HP>=200,'HP']=180
```

```python
toyota2_df.describe()
```

```python
toyota2_df.loc[toyota2_df.Age_08_04>=12,'Old_Cars']='Old'
```

```python
toyota2_df.loc[toyota2_df.Age_08_04<12,'Old_Cars']='Not_too_old'
```

```python
toyota2_df.loc[toyota2_df.Fuel_Type=='Diesel','Diesel']=1
toyota2_df.loc[toyota2_df.Diesel.isnull(),'Diesel']=0
```

```python
toyota2_df.loc[toyota2_df.Fuel_Type=='CNG','CNG']=1
toyota2_df.loc[toyota2_df.CNG.isnull(),'CNG']=0
```

```python
toyota2_df.loc[toyota2_df.Fuel_Type=='Petrol','Petrol']=1
toyota2_df.loc[toyota2_df.Petrol.isnull(),'Petrol']=0
```
